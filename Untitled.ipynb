```python
pre_img = cv2.imread('pd_subPartlist_dw40f.png')
image = imutils.resize(pre_img, width=1300)

rgb_planes = cv2.split(image)

result_planes = []
result_norm_planes = []
for plane in rgb_planes:
    dilated_img = cv2.dilate(plane, np.ones((7,7), np.uint8))
    bg_img = cv2.medianBlur(dilated_img, 21)
    diff_img = 255 - cv2.absdiff(plane, bg_img)
    norm_img = cv2.normalize(diff_img,None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)
    result_planes.append(diff_img)
    result_norm_planes.append(norm_img)

result = cv2.merge(result_planes)
result_norm = cv2.merge(result_norm_planes)

gray = cv2.cvtColor(result_norm, cv2.COLOR_BGR2GRAY)
gary = cv2.GaussianBlur(gray, (5, 5), 0)

se=cv2.getStructuringElement(cv2.MORPH_RECT , (1,1))
gray = cv2.morphologyEx(gray, cv2.MORPH_DILATE, se)

im_bw = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)[1]
# img_bw = 255*(cv2.cvtColor(result_norm, cv2.COLOR_BGR2GRAY) > 1).astype('uint8')
# im_bw = cv2.threshold(im_bw, 190, 255, cv2.THRESH_BINARY)[1]

# noiseless_image_bw = cv2.fastNlMeansDenoising(im_bw, None, 10, 10, 10)

# kernel = np.ones((1,1),np.uint8)
# # erosion = cv2.erode(noiseless_image_bw,kernel,iterations = 1)
# # opening = cv2.morphologyEx(noiseless_image_bw, cv2.MORPH_OPEN, kernel)
# dilation = cv2.dilate(noiseless_image_bw,kernel,iterations = 1)

# sharpen_kernel = np.array([[-1,-1,-1], [-1,100,-1], [-1,-1,-1]])
# sharpen = cv2.filter2D(dilation, -1, sharpen_kernel)

# cleaned = morphology.remove_small_objects(im_bw, min_size=10000)

plt.figure(figsize = (15, 10))
plt.imshow(cv2.cvtColor(im_bw, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
d = pytesseract.image_to_data(im_bw, lang='eng', config='--psm 11 --oem 3 -c preserve_interword_spaces=1 -c tessedit_char_whitelist="0123456789 "', output_type=Output.DICT)
n_boxes = len(d['level'])

rc = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

int_list = d['text']

location_list = {'int': [], 'location': []}

for i in range(n_boxes):
    if int_list[i] != '':
        (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
        cv2.rectangle(rc, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(rc, int_list[i], (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (0, 0, 0), 2)
        
        location_list['int'].append(int_list[i])
        location_list['location'].append((x, y, w, h))
#     print((x, y, w, h))

plt.figure(figsize = (15, 10))
plt.imshow(rc)
plt.show()

pd.DataFrame.from_dict(location_list)
```
